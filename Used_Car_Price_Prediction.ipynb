In the dataset, there are multiple cars, some of which have a count greater than one. The resale value of a car often depends on its manufacturer. To enhance our analysis, we will extract the manufacturer name from the 'Name' column and incorporate it as a new column in the dataset.

```python
manufacturer = df['Name'].str.split(' ',expand=True)
df['Manufacturer'] = manufacturer[0]
```

The 'Year' column in the dataset lacks intrinsic significance when considered independently. To enhance its utility, we will calculate the age of each car based on the current year and incorporate this information into the dataset.

```python
df.drop(['Name','Year'],axis=1,inplace=True)
```

```python
df.head()
```

The 'Mileage' column indicates the mileage of the car, a factor that influences its sales price. Therefore, we will extract the numerical values from this column. To address the missing values, we will fill them with the mean mileage and update the column accordingly.

```python
mileage = df['Mileage'].str.split(' ',expand=True)
df['Mileage'] = pd.to_numeric(mileage[0],errors='coerce')
df['Mileage'].fillna(df['Mileage'].astype('float').mean(),inplace=True)
```

The 'Engine' column represents the engine capacity in cubic centimeters (CC), which impacts the car's sales price. We will remove the "CC" notation and extract the numerical values from this column. To handle the missing values, we will fill them with the mean engine capacity and update the column accordingly.

```python
engine = df['Engine'].str.split(' ',expand=True)
df['Engine']=pd.to_numeric(engine[0],errors='coerce')
```

```python
df['Engine'].unique()
```

```python
df['Engine'].fillna(df['Engine'].astype('float').mean(),inplace=True)
```

The 'Power' column indicates the brake horsepower (bhp) of each car, which influences its resale price. To prepare this data for analysis, we will remove the "bhp" suffix from the entries and extract the numerical values. Additionally, as there are missing values in this column, we will fill these gaps with the mean bhp and update the column accordingly.

```python
power = df['Power'].str.split(' ',expand=True)
df['Power']=pd.to_numeric(power[0],errors='coerce')
df['Power'].fillna(df['Power'].astype('float').mean(),inplace=True)
```

Since the 'Seats' column is a discrete variable and has some missing values, we will fill them with the mode value of the same column.

```python
df['Seats'].mean()
```

```python
df['Seats'].mode()
```

```python
df['Seats'].fillna(df['Seats'].astype('float').mode()[0],inplace=True)
```

After filling in the missing values with the appropriate values, we will conduct a final check to identify any remaining null values in the dataset.

# 6. Preprocessing

**Variable Assignment for Analysis**

In this analysis, we will assign our dependent and independent variables as follows:

#### Dependent Variable
- **Price**: This is the target variable we aim to predict.

#### Independent Variables
The following features will be used as predictors in our model:
- **Manufacturer**: The brand of the car.
- **Year Used**: The number of years since the car was manufactured.
- **Kilometers Driven**: The total distance the car has traveled.
- **Mileage**: The fuel efficiency of the car.
- **Engine**: The engine capacity of the car in cubic centimeters.
- **Power**: The brake horsepower (bhp) of the car.
- **Seats**: The number of seats in the car.

```python
# Assigning the independent variables
x = df.drop(['Price'],axis=1)
# Assigning the dependent variable
y = df['Price']
```

To evaluate the performance of our predictive model, we will divide the dependent and independent variables into training and testing datasets using the train-test split method. This process allows us to train the model on one subset of the data while testing its accuracy on a separate subset.

```python
# Split the data into training and testing sets
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=45)
```

**Creating Dummy Variables**

To incorporate categorical variables into our predictive model, we will create dummy columns for all categorical features in the training independent dataset. This includes the following columns: `Manufacturer`, `Fuel_Type`, `Transmission`, and `Owner_Type`. 

```python
# Create dummy variables for categorical columns in the training independent dataset
x_train = pd.get_dummies(x_train,columns=['Manufacturer','Fuel_Type','Transmission','Owner_Type'],drop_first=True)
# Obtain the dimensions of the modified training independent dataset
print(x_train.shape)
```

Creating Dummy Variables

To incorporate categorical variables into our predictive model, we will create dummy columns for all categorical features in the testing dataset. This includes the following columns: Manufacturer, Fuel_Type, Transmission, and Owner_Type.

```python
# Create dummy variables for categorical columns in the testing dataset
x_test = pd.get_dummies(x_test,columns=['Manufacturer','Fuel_Type','Transmission','Owner_Type'],drop_first=True)
# Obtain the dimensions of the modified training independent dataset
print(x_test.shape)
```

**Addressing Dimensional Discrepancies**

We have observed that the process of creating dummy columns results in dimensional discrepancies between the training and test independent datasets. To ensure consistency between these datasets, we will fill in any missing dummy columns in the test independent dataset with zeros.

```python
# Identify missing columns in the test independent dataset
miss_col = set(x_train.columns)-set(x_test.columns)
# Add missing columns to the test independent dataset with default value 0
for col in miss_col:
    x_test[col]=0
x_test = x_test[x_train.columns]
```

```python
# Verify the dimensions of both datasets
print('X_train',x_train.shape)
print('X_test',x_test.shape)
```

After the modifications, the dimensions of the columns in both the training and test independent datasets are now identical.

We are standardizing the independent variables in the dataset. This means we will scale the data so that it has a mean of 0 and a standard deviation of 1.

```python
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)
```

# 7. Model Fitting
<a href=#cont>Back to Table of Contents</a>

We are defining a function to implement machine learning algorithms using GridSearchCV. This function will predict the dependent variable by fitting the specified model and create a pickle file for the model using the given algorithm name.

Furthermore, it will provide details such as the algorithm name, best parameters, R² score in percentage format, mean absolute error, mean squared error, and root mean squared error by comparing the predicted values with the dependent test dataset.

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, ExtraTreesRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
import xgboost as xgb
import catboost as cb

# Initialize models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Extra Trees': ExtraTreesRegressor(random_state=42),
    'XGBoost': xgb.XGBRegressor(random_state=42),
    'CatBoost': cb.CatBoostRegressor(verbose=0, random_state=42)
}

# Create a dictionary to store the evaluation metrics for each model
model_performance = {}

# Train and evaluate each model
for name, model in models.items():
    # Fit the model
    model.fit(x_train, y_train)
    
    # Make predictions
    y_pred = model.predict(x_test)
    
    # Calculate evaluation metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = sqrt(mse)
    
    # Store the results
    model_performance[name] = {
        'R2 Score': r2 * 100,
        'Mean Absolute Error': mae,
        'Mean Squared Error': mse,
        'Root Mean Squared Error': rmse
    }

model_performance
```

- **CatBoost** and **Extra Trees** performed the best, both achieving R² scores above 91%.
- **Random Forest** and **XGBoost** also performed well, with **Random Forest** having a lower Root Mean Squared Error (RMSE) compared to **XGBoost**.
- **Linear Regression**, **Ridge**, and **Lasso** models performed less effectively, with lower R² scores and higher errors compared to the ensemble models.

```python
from sklearn.model_selection import GridSearchCV
import catboost as cb
from sklearn.ensemble import ExtraTreesRegressor

# Define parameter grids for CatBoost and Extra Trees
param_grid_catboost = {
    'depth': [4, 6, 8],
    'learning_rate': [0.01, 0.1, 0.5],
    'iterations': [100, 200]
}

param_grid_extra_trees = {
    'n_estimators': [50, 100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'min_samples_split': [2, 5, 10]
}

# Initialize models for tuning
catboost_model = cb.CatBoostRegressor(random_state=42, verbose=0)
extra_trees_model = ExtraTreesRegressor(random_state=42)

# Perform Grid Search for CatBoost
grid_search_catboost = GridSearchCV(estimator=catboost_model, param_grid=param_grid_catboost, 
                                     scoring='r2', cv=3, n_jobs=-1)
grid_search_catboost.fit(x_train, y_train)

# Perform Grid Search for Extra Trees
grid_search_extra_trees = GridSearchCV(estimator=extra_trees_model, param_grid=param_grid_extra_trees, 
                                        scoring='r2', cv=3, n_jobs=-1)
grid_search_extra_trees.fit(x_train, y_train)

# Get best parameters and scores
best_catboost_params = grid_search_catboost.best_params_
best_catboost_score = grid_search_catboost.best_score_

best_extra_trees_params = grid_search_extra_trees.best_params_
best_extra_trees_score = grid_search_extra_trees.best_score_

print("Best CatBoost Parameters:", best_catboost_params)
print("Best CatBoost Score:", best_catboost_score)

print("Best Extra Trees Parameters:", best_extra_trees_params)
print("Best Extra Trees Score:", best_extra_trees_score)
```

```python
from catboost import CatBoostRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

# Train CatBoost with best parameters
catboost_model_final = CatBoostRegressor(
    depth=6,
    iterations=200,
    learning_rate=0.1,
    random_state=42,
    verbose=0
)

catboost_model_final.fit(x_train, y_train)

# Predict on the test set
catboost_y_pred = catboost_model_final.predict(x_test)

# Evaluate performance
catboost_r2 = r2_score(y_test, catboost_y_pred)
catboost_mae = mean_absolute_error(y_test, catboost_y_pred)
catboost_mse = mean_squared_error(y_test, catboost_y_pred)
catboost_rmse = np.sqrt(catboost_mse)

# Train Extra Trees with best parameters
extra_trees_model_final = ExtraTreesRegressor(
    max_features='sqrt',
    min_samples_split=2,
    n_estimators=200,
    random_state=42
)

extra_trees_model_final.fit(x_train, y_train)

# Predict on the test set
extra_trees_y_pred = extra_trees_model_final.predict(x_test)

# Evaluate performance
extra_trees_r2 = r2_score(y_test, extra_trees_y_pred)
extra_trees_mae = mean_absolute_error(y_test, extra_trees_y_pred)
extra_trees_mse = mean_squared_error(y_test, extra_trees_y_pred)
extra_trees_rmse = np.sqrt(extra_trees_mse)

# Print performance metrics
print("CatBoost Performance:")
print(f"R² Score: {catboost_r2 * 100:.2f}%")
print(f"Mean Absolute Error: {catboost_mae:.2f}")
print(f"Mean Squared Error: {catboost_mse:.2f}")
print(f"Root Mean Squared Error: {catboost_rmse:.2f}")

print("\nExtra Trees Performance:")
print(f"R² Score: {extra_trees_r2 * 100:.2f}%")
print(f"Mean Absolute Error: {extra_trees_mae:.2f}")
print(f"Mean Squared Error: {extra_trees_mse:.2f}")
print(f"Root Mean Squared Error: {extra_trees_rmse:.2f}")
```

```python
# Load the model
with open('catboost_model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)
```

After loading the model, you can make predictions on new data.

```python
#Making predictions on test data
predictions = loaded_model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate RMSE
rmse = mean_squared_error(y_test, predictions, squared=False)
print(f"RMSE: {rmse}")

# Evaluate R² score
r2 = r2_score(y_test, predictions)
print(f"R² Score: {r2}")
```

- RMSE (Root Mean Squared Error): 3.42, which suggests that on average, the model's predictions are off by about 3.42 units from the actual values.

- R² Score: 0.91, which indicates that approximately 91% of the variance in the target variable is explained by the model. This is typically considered a strong R² score in regression analysis.
